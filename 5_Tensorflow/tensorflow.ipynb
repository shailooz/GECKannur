```python
#from tensorflow import keras
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Input,Dense,Add,Concatenate

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :]  # we only take the first two features.
y = iris.target
encoder=LabelBinarizer()
Y=encoder.fit_transform(y)
#print(X)
#print(Y)
```

```python
#y = y.reshape((len(y),-1))
#encoder = OneHotEncoder(sparse=False)
#Y = encoder.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
tf.random.set_seed(42)
'''
model = Sequential([                  
 Input((4)),
 Dense(100, activation='relu'),
 Dense(50, activation='relu'),
 Dense(3, activation='softmax')
])

model = Sequential()
model.add(Dense(100, activation='relu', input_shape = (4,)))
model.add(Dense(50, activation='relu'))
model.add(Dense(3, activation='softmax'))
'''

input = Input(shape=(4,))
x1 = Dense(100, activation='relu', input_shape = (4,))(input)
x = Dense(50, activation='relu')(x1)
xx1 = Concatenate()([x,x1])
output = Dense(3, activation='softmax')(xx1)
model = Model(input, output )


model.compile(optimizer='sgd',loss="categorical_crossentropy",metrics=['accuracy'])

model.summary()

tf.keras.utils.plot_model(model,show_shapes=True)
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=16)
model.evaluate(X_test,y_test)
```

```python
y_pred  = np.argmax(model(X_test), axis =-1)
print(y_pred)

y_true =  np.argmax(y_test, axis =-1)
print(y_true)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_true, y_pred)
print(cm)
cr = classification_report(y_true, y_pred)
print(cr)
```

```python
import seaborn as sns
sns.heatmap(cm,cmap='Greens', annot=True,xticklabels=iris.target_names,yticklabels=iris.target_names)
```
